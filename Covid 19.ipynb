Data visualization to show the no of Patient according to state

```python
import matplotlib.pyplot as plt
import pandas as pd
cdata=pd.read_excel("Covid_19.xlsx")

dt=cdata.groupby(['Detected State','Gender']).size().reset_index(name="Count")
fig, ax1 = plt.subplots(figsize = (10,10)) 
ax1.barh(dt["Detected State"],dt['Count'],color="#ff884d")
plt.ylabel("States")
plt.xlabel("No. of Patients")
plt.title("Total no. of Patients according to State")
plt.tight_layout()
plt.show()
```

The graph Shows the no of Male and Female Among the patients

```python

val=cdata.groupby(["Detected State","Gender"])["Gender"].count()
val.unstack().plot.bar(figsize=[10,10],stacked=True,color=["#33adff","#4dff4d"])
plt.title("Total no. of Patients affected according to Gender and State ")
plt.xlabel("States")
plt.ylabel("No.of Patients")
plt.tight_layout()
plt.show()
```

The graph shows the no of patent according to the Date

```python
dt=cdata["Date Announced"].value_counts()
dt.plot.line(figsize=(13,13),linewidth="5")
plt.xlabel("Date")
plt.ylabel("No.of Patients")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
cdata=pd.read_excel("Covid_19.xlsx")

a=cdata.groupby(['Detected State']).size().reset_index(name="Count")
State = a["Detected State"]
Patient = a["Count"]
fig, ax1 = plt.subplots(figsize = (24,12)) 
explode = (0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.500,0.900,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05)
ax1.pie(Patient, labels=State,autopct='%1.1f%%', shadow=False, startangle=140,explode=explode)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Percentage of people affected according State', fontsize=25)
plt.tight_layout()
plt.show()
```

No of patent accoding to the Directions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter


df = pd.read_excel (r'Covid_19.xlsx')
stt=df['Detected State']

north={"Rajastan","Punjab","Delhi","Arunachal Pradesh","Uttarakhand","Haryana","Jammu and Kashmir","Manipur""Uttar Pradesh","Ladakh","Chandigarh"}
nt=0
south={"Kerla","Telangana","Tamil Nadu","Karnataka","Puducherry","Andhra Pradesh"}
st=0
west={"Maharashtra","Gujarat"}
wt=0
east={"Odisha","Bihar"}
et=0
for x in stt:
    if x in north:
        nt+=1
    elif x in south:
        st+=1
    elif x in west:
        wt+=1
    elif x in east:
        et+=1

#stts={"North","South","East","West"}
plt.bar("North",nt)
plt.bar("South",st)
plt.bar("East",et)
plt.bar("West",wt)
plt.title("State wise cases ")
plt.xlabel("States")
plt.ylabel("No. of Patients")
plt.show()
```
